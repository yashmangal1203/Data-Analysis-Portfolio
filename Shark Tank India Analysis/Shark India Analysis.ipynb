# Shark Tank India (Season 1) Exploratory Data Analysis (EDA)

 Dataset - https://www.kaggle.com/thirumani/shark-tank-india

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 50)
```

```python
shark_tank = pd.read_csv('Shark Tank India.csv')
```

```python
print('Total Rows in dataset: ',shark_tank.shape[0])
print('Total Columns in dataset: ',shark_tank.shape[1])
```

```python
shark_tank.head()
```

```python
shark_tank.tail().T
```

```python
# Chaning datatypes of those required.

shark_tank['Male Presenter'] = shark_tank['Male Presenter'].astype(pd.Int32Dtype())
shark_tank['Female Presenter'] = shark_tank['Female Presenter'].astype(pd.Int32Dtype())
shark_tank['Started in'] = shark_tank['Started in'].astype(pd.Int32Dtype())
shark_tank['Yearly Revenue'] = shark_tank['Yearly Revenue'].astype(pd.Int32Dtype())
shark_tank['Monthly Sales'] = shark_tank['Monthly Sales'].astype(pd.Int32Dtype())
shark_tank['Valuation Offered'] = shark_tank['Valuation Offered'].astype(pd.Int32Dtype())
```

## Some Statistics of the entire season 1 of Shark Tank India broadcasted on Sony Channel and available for streaming on SonyLiv

 ### Total number of Pitches in the whole season

```python
print(f"Season : {shark_tank['Season Number'].max()}")
print(f"Total Episodes : {shark_tank['Episode Number'].max()}")
print(f"Total number of pitches whole season : {shark_tank['Pitch Number'].max()}")
```

### Types on Industries that pitched to the sharks

```python
industries = shark_tank['Industry'].value_counts()
print(industries)

plt.figure(figsize=(20,7),dpi=100)

colors = ['#001219','#005F73','#0A9396','#94D2BD','#E9D8A6','#EE9B00','#CA6702','#BB3E03','#AE2012','#9B2226']
barchart = plt.bar(industries.index,height=industries.values,color=colors)

for rect, label in zip(barchart.patches, industries.values):
    height = rect.get_height()
    plt.text(
        rect.get_x() + rect.get_width() / 2, height+0.1, label, ha="center", va="bottom"
    )


plt.xlabel("Industries")
plt.ylabel("Number of startups")

plt.show()
```

### Gender wise statistics of the pitchers

```python
total_presenters = shark_tank['Number of Presenters'].sum()
total_male_presenter = shark_tank['Male Presenter'].sum()
total_female_presenter = shark_tank['Female Presenter'].sum()
couple_presenter = shark_tank['Couple Presenters'].sum()


print(f"Total pitchers for all pitches : {total_presenters} \n")
print(f"Total Male pitchers : {total_male_presenter}")
print(f"Total Female pitchers : {total_female_presenter}")
print(f"Total Couple pitchers : {couple_presenter} \n")
print(f"Percentage of Male pitchers to total pitchers : {100*total_male_presenter/total_presenters:.2f}%")
print(f"Percentage of Female pitchers to total pitchers : {100*total_female_presenter/total_presenters:.2f}%")
print(f"Percentage of Couple pitchers to total pitchers : {100*couple_presenter/total_presenters:.2f}%")
```

### Total offers received and not received.

```python
received_offer_yes = shark_tank['Received Offer'].value_counts()[1]
received_offer_no = shark_tank['Received Offer'].value_counts()[0]
total_offers = shark_tank['Pitch Number'].max()

print(F"Total offers: {total_offers}")
print(F"Offers received: {received_offer_yes}")
print(F"Offers not received: {received_offer_no}")
print(f"That is {100*received_offer_yes/total_offers:.2f}% of the total pitches received offers and {100*received_offer_no/total_offers:.2f}% did not.")

plt.bar(shark_tank['Received Offer'].value_counts().index,height=shark_tank['Received Offer'].value_counts().values,color=['#252422','#eb5e28'])

plt.xticks(shark_tank['Received Offer'].value_counts().index)
plt.xlabel('Received Offer')
plt.ylabel('Number of startups')

plt.show()
```

### How many startups accepted the offers and how many rejected

```python
shark_tank['Accepted Offer'].value_counts()

print(f"Total offers received: {received_offer_yes}")
print(f"Offers Accepted: {shark_tank['Accepted Offer'].value_counts().iloc[0]}")
print(f"Offers Rejected: {shark_tank['Accepted Offer'].value_counts().iloc[1]}")
print(f"That is of the {received_offer_yes} offers received, {100*shark_tank['Accepted Offer'].value_counts(normalize=True).iloc[0]:.2f}% were accepted and {100*shark_tank['Accepted Offer'].value_counts(normalize=True).iloc[1]:.2f}% were rejected.")

plt.bar(shark_tank['Accepted Offer'].value_counts().index,height=shark_tank['Accepted Offer'].value_counts().values,color=['#e09f3e','#9e2a2b'])

plt.xticks(shark_tank['Accepted Offer'].value_counts().index)
plt.xlabel('Offers Rejected/Accepted')
plt.ylabel('Number of Offers')

plt.show()
```

### Top 10 investments in startups in Shark Tank

```python
top_10_startups = shark_tank.iloc[shark_tank['Total Deal Amount'].nlargest(10).index][['Startup Name','Total Deal Amount']]
top_10_startups = top_10_startups.set_index(['Startup Name']).sort_index()
print(top_10_startups)

plt.figure(figsize=(20,8))

colors = ['#f94144','#f3722c','#f8961e','#f9844a','#f9c74f','#90be6d','#43aa8b','#4d908e','#577590','#277da1']
plt.bar(top_10_startups.index,height=top_10_startups.values.flatten(),color=colors)
plt.xticks(top_10_startups.index)

plt.xlabel('Startup Name',fontdict={'fontsize':15})
plt.ylabel('Amount Invested in Rs.',fontdict={'fontsize':15})
plt.title('Amount invested in top 10 startups',fontdict={'fontsize':20})

for i, val in enumerate(top_10_startups.values.flatten()):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})


plt.show()
```

### Amounts invested by each shark ranked from highest to lowest

```python
sharks_ranked = shark_tank[['Ashneer Investment Amount','Namita Investment Amount','Anupam Investment Amount','Vineeta Investment Amount','Aman Investment Amount','Peyush Investment Amount','Ghazal Investment Amount']].sum().sort_values(ascending=False)
print((sharks_ranked/100).round(2).to_string())

plt.figure(figsize=(3,3),dpi=150)

colors = ['#FCDE9C','#FFA552','#BA5624','#381D2A','#C4D6B0','#52D1DC','#475B5A']
plt.pie(sharks_ranked,autopct='%1.1f%%',colors=colors,labels=sharks_ranked.index.str[:-18],textprops={'fontsize':'xx-small'})

plt.show()
```

### Amount of equity invested by each Shark

```python
sharks_ranked = shark_tank[['Ashneer Investment Equity','Namita Investment Equity','Anupam Investment Equity','Vineeta Investment Equity','Aman Investment Equity','Peyush Investment Equity','Ghazal Investment Equity']].sum().sort_values(ascending=False)
print((sharks_ranked).round(2).to_string())
```

### Amount of debt given by each Shark

```python
sharks_ranked = shark_tank[['Ashneer Debt Amount','Namita Debt Amount','Anupam Debt Amount','Vineeta Debt Amount','Aman Debt Amount','Peyush Debt Amount','Ghazal Debt Amount']].sum().sort_values(ascending=False)
print((sharks_ranked).round(2).to_string())
```

### Total Investments by Ashneer

```python
print(f"Total investments by Ashneer: {shark_tank[shark_tank['Ashneer Investment Amount'] > 0]['Ashneer Investment Amount'].count()}")

print(f"Total Amount invested by Ashneer: {shark_tank[shark_tank['Ashneer Investment Amount'] > 0]['Ashneer Investment Amount'].sum()/100:.2f} Crores")

print()
print('*'*75)

print("\nDifference startups and industries in which Ashneer has invested:")
ashneer_investments = shark_tank[shark_tank['Ashneer Investment Amount'] > 0][['Startup Name','Industry','Ashneer Investment Amount']].set_index('Startup Name').sort_values('Ashneer Investment Amount',ascending=False)
print(ashneer_investments)

print()
print('*'*75)
print()

print(f"Ashneer's Industry choices:\n{ashneer_investments['Industry'].value_counts().to_string()}")

plt.figure(figsize=(3,3),dpi=100)
plt.pie(ashneer_investments['Industry'].value_counts().values,labels=ashneer_investments['Industry'].value_counts().index,autopct='%1.1f%%')
plt.title('Industry Allocation of Ashneer')
plt.show()
```
